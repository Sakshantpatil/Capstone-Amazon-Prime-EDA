### 3.3 Missing Value Percentage Analysis

Missing value analysis is an essential step in data preprocessing as it helps identify incomplete or unavailable information within the dataset. Understanding the distribution of missing values enables appropriate handling techniques and ensures reliable and accurate data analysis.

````python
# Missing Values Summary
# Missing Values percentage

missing_summary = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Missing Percentage': (df.isnull().sum() / len(df)) * 100
})

# Display only columns with missing values
missing_summary = missing_summary[missing_summary['Missing Values'] > 0]

# Round percentage to 2 decimal places
missing_summary['Missing Percentage'] = missing_summary['Missing Percentage'].round(2)

# Sort by highest missing percentage
missing_summary.sort_values(by='Missing Percentage', ascending=False)
````

### 3.4 Visualize Missing values

````python
#visualize Missing values

plt.figure(figsize=(10,6))

missing_summary.sort_values(
    by='Missing Values'
)['Missing Values'].plot(
    kind='barh',
    color='skyblue'
)

plt.title('Missing Values by Column')
plt.xlabel('Number of Missing Values')
plt.ylabel('Columns')

plt.show()
````

#### Observation



The missing value analysis shows that most columns have a relatively low percentage of missing values. However, the **seasons** column has approximately **86.25%** missing values because it applies only to TV Shows, while Movies naturally do not have season information.

## 3.5 Feature Engineering


Feature Engineering is the process of creating new features or transforming existing ones to make the dataset more meaningful and easier to analyze. In this project, new features are created from the existing data to simplify visualizations and generate more actionable business insights.

###1. Creating Content Age():

````python
# Write your code to make your dataset analysis ready.
# Feature Engineering- Creating Content Age

current_year = 2026

df['content_age'] = current_year - df['release_year']

df[['title', 'release_year', 'content_age']].head()
````

### Observation



A new feature named **content_age** was created by calculating the difference between the current year and the release year of each title. This feature helps analyze the age distribution of Amazon Prime content and supports further business insights related to content trends.

### 2. Creating Primary Genre():

````python
# Feature Engineering - Creating Primary Genre
import ast

df['primary_genre'] = df['genres'].apply(
    lambda x: ast.literal_eval(x)[0] if pd.notnull(x) and len(ast.literal_eval(x)) > 0 else None
)

df[['genres', 'primary_genre']].head()
````

### Observation


A new feature named **primary_genre** was created by extracting the first listed genre from each title. This simplifies genre-based analysis and enables clear comparisons across different categories in the visualization phase.

### 3. Creating Primary Country():




````python
# Feature Engineering - Creating Primary Country

df['primary_country'] = df['production_countries'].apply(
    lambda x: ast.literal_eval(x)[0] if pd.notnull(x) and len(ast.literal_eval(x)) > 0 else None
)

df[['production_countries', 'primary_country']].head()
````

### Observation


The **primary_country** feature was created by extracting the first listed production country for each title. This enables country-wise analysis of Amazon Prime content and helps identify major content-producing regions.

### Chart - 17 Top Directors vs Average IMDb Score (Merged)

#### **Summary**

This chart analyzes the performance of **Top Directors** by comparing their **average IMDb scores** using the **merged Titles and Credits datasets**. It helps identify directors who have consistently produced highly rated content while directing multiple titles.

#### Step 1: Use the merged dataset to prepare the data for the chart.

````python
# Keep only directors
directors = merged_df[merged_df['role'] == 'DIRECTOR']

# Calculate director statistics
director_summary = (
    directors
    .dropna(subset=['imdb_score'])
    .groupby('name')
    .agg(
        Number_of_Titles=('id', 'count'),
        Average_IMDb=('imdb_score', 'mean')
    )
    .query("Number_of_Titles >= 5")
    .sort_values(
        ['Average_IMDb', 'Number_of_Titles'],
        ascending=[False, False]
    )
    .head(10)
)

director_summary
````

####Step 2: Visualization

````python
# Chart 17: Top Directors by Average IMDb Score (Minimum 5 Titles)

plt.figure(figsize=(10,6), facecolor="#0F171E")

ax = sns.barplot(
    data=director_summary.reset_index(),
    x="Average_IMDb",
    y="name",
    palette="Blues_r"
)

ax.set_facecolor("#0F171E")

plt.title(
    "Top Directors by Average IMDb Score (Minimum 5 Titles)",
    fontsize=16,
    fontweight="bold",
    color="white"
)

plt.xlabel("Average IMDb Score", fontsize=12, color="white")
plt.ylabel("Director", fontsize=12, color="white")

plt.xticks(color="white")
plt.yticks(color="white")

plt.grid(axis='x', linestyle='--', alpha=0.2, color='white')

# Value labels
for i, value in enumerate(director_summary["Average_IMDb"]):
    plt.text(
        value + 0.02,
        i,
        f"{value:.2f}",
        va="center",
        fontsize=10,
        color="white"
    )

plt.show()
````

##### 1. Why did you pick the specific chart?

A horizontal bar chart was selected to compare the average IMDb scores of directors with at least five titles. This makes it easy to rank directors while ensuring the comparison is based on a sufficient number of productions.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that **Tony Charmoli achieved the highest average IMDb score (8.14)** among directors with at least five titles, followed by** Hideaki Anno (7.50) and William Wyler (7.47).** This indicates that these directors consistently produced highly rated content rather than relying on the success of a single title.
